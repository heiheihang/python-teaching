## Stack

In a stack, it is essentially a list but you only have direct access to its last element. 

I can perform the following operation to a stack `s`:


*   `s.append(n)`: I can add a new element to the end of the stack
*   `s.pop(-1)`: I can remove the last element of the stack
*   `s[-1]` I can view the last element of the stack

In Python (and many other languages), of course we can do `pop(b)` and pop away any element in a list, or we can do `s[b]` to view any element in a list. However, it is important to keep this data structure in mind as many intermediate problems will make use of stack for fast operations.

# Queue
Queue is similar to a stack, but you can only add element from the start of the list and pop element from the end of the list. 

I can perform the following operation to a queue `q`:


*   q.pop(-1)
*   q.insert(0,n)
*   q[-1]

# Dictionary

Dictionary is like a list, with index replaced by values. You may have a dictionary:


```
d = {'one' : 1, 'two' : 2}
d['one'] = 1
d['two'] = 2
```
To initialise an empty dictionary (dict), we do: `d = {}`

To add a new entry to the dictionary, we can:

```
d['one'] = 1
```
or

```
if(n not in d):
  d[n] = 1
else:
  d[n] += 1
```
