## Stack


*   Can only insert element at the "head" of the stack `s.append(n)`
*   Can only pop the element at "head" `s.pop(-1)`
*   Can only look at the last element `s[-1]`




````
stack = [2,4,5,6,7,5,4,2,3,4]

stack.append(9)

print(stack) #[2,4,5,6,7,5,4,2,3,4,9]

stack.pop(-1)

print(stack) #[2,4,5,6,7,5,4,2,3,4]

print(stack[-1]) #4

````

## Queue



*   Can only insert element in the front of the queue
*   Can only pop element at the end of the queue
*   Can only view element at the end of the queue



````
queue = [2,4,5,6,7,5,4,2,3,4]

queue.insert(0,3)

print(queue) #[3, 2, 4, 5, 6, 7, 5, 4, 2, 3, 4]

queue.pop(-1)

print(queue) #[3, 2, 4, 5, 6, 7, 5, 4, 2, 3]

print(queue[-1]) #3
````

## Dict



*   Initilised with {}
*   Holds `key : value` pairs `{'key_1': value_1, 'key_2': value_2}`
*   Insert new `key : value` pair with `d[new_key] = new_value`



````
d = {}

d[1] = 2
d[2] = 1

print(d)
````
