## Range and Loop

Range describes a sequence of numbers, and the most common type of range we will be using is something like (0,1,2,3,4,...,n).

For example, range(5) contains elements: 0,1,2,3,4 

Notice that it starts from 0 and ends in 4, and it contains 5 elements.



This is an example of the range 0 to 9. Notice the use of list, which will be explained next week! 

````
print(list(range(0,10)))
````

An example of loop with range!

````
for i in range(5):
  print(i)
  print("Hello World!")
````

Another example:

````
for i in range(5):
  print(i+1)
````

Another example:

````
n = 2
for i in range(5):
  n = n + 2
  print(n)
````

Another example:


````
n = 2
for i in range(5):
  n = n * 2
print(n)
````

````
n = 5
n = n * 3 + n

#(new value of n) = (old value of n = 5) * 3 + (old value=5)
````

A more complicated example:

````
n = 3
for i in range(4):
  if(n %2 == 0):
    n = n * 2
  else:
    n = n * 4
````
