# Task 2 Part 1

Compute 1+2+3+4+...+100 using a for loop

```
#Your Code Here
x = 0
huge_sum = 0

for i in range(5):
  x = x + 1
  huge_sum = huge_sum + x #new value = old value + x
  print(x, huge_sum)


# huge_sum = 0 (huge_sum) + 1 (x) = 1
# huge_sum = 1 (huge_sum) + 2 (x) = 3
# huge_sum = 3 (huge_sum) + 3 (x) = 6
# huge_sum = 6 (huge_sum) + 4 (x) = 10
# huge_sum = 10 (huge_sum) + 5(x) = 15
# huge_sum = 1 + 2 + 3 + 4 + .. + 100
```
