# Week 4: For-Loop Homework



You have learnt quite a lot about for-loop! Let's try to remind yourself what it does!

```
n = 0
for i in range(10):
  print(n)
  n = n + 1
```

We can put conditionals (if statements) within a for loop as well!

```
n = 0

for i in range(5):
  print(n)
  if(n % 2 == 0):
    print("even!")
    n = n + 3
  else:
    print("odd!")
    n = n + 3
```

We can also put a for loop within a for loop!

```
for i in range(5):
  for j in range(5):
    print(i,j)
```
