# Revision

You have learn else-if (elif) that can be used in the following way

````
x = 5

if(x % 2 == 0):
  print("x is even")
elif(x > 3):
  print("x is larger than 3")
else:
  print("x is others")
````

You have also learnt AND/OR

````
x = 12

if(x > 5 or x < 2):
  print("x is not between 2 and 5")

if(x % 3 == 0 and x > 9):
  print("x is divisible by 3 and larger than 9")
````

Finally, you have lerant range, and while/for loops.

````
x = 0
for i in range(5):
  x = x + 1 
  print(x)

i = 0
x = 0
while(i < 5):
  x += 1
  i += 1
  print(x)

````
