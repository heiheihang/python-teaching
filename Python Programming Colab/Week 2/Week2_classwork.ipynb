# If-else, elif

Remember that we learnt if-else in the following form:

```
if(x == 0):
  print("x is 0")
else:
  print("x is not 0")
```

Now we want the if-else branch to not only recognize 0, but also recognize 1. We introduce a new idea: else-if. Else-if is denoted as elif. Look at the example below:



```
if(x == 0):
  print("x is 0")
elif( x == 1):
  print("x is 1")
else:
  print("x is not 0 nor 1")
```

# And Or

If we want a conditional branch to hold 2 conditions, we need to use and/or.

For example if we want to check if x is even and x is not 0, we can do the following:


```
if(x % 2 == 0 and x != 0):
  print("x is even and is not 0")
else:
  print("x is odd or is 0")
```

We can also use OR when we want to enter a branch if either one of the conditions is satisfied. 

For example if we want to check if x is odd or x is larger than 10, we can do the following:


```
x = 5
if(x % 2 == 1 and x > 10):
  print("x is odd or x is larger than 10")
else:
  print("x is even and is smaller than or equal to 10")
```

# Range and Loop

Range describes a sequence of numbers, and the most common type of range we will be using is something like (0,1,2,3,4,...,n).

For example, range(5) contains elements: 0,1,2,3,4 

Notice that it starts from 0 and ends in 4, and it contains 5 elements.

This is an example of the range 0 to 9. Notice the use of list, which will be explained next week! 

````
print(list(range(0,10)))
````

We can repeat the same process several times with loops.

For example, if we want to multiply x by 2 5 times, we can do the following:


```
x = 1
for i in range(5):
  x = x * 2
  print(x)
```



````
x = 1

for i in range(5):
  x = x * 2
  print(x)
````
