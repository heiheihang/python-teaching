We have also learnt modulo, which simply gives the remainder of a division. 

For example:

10 % 3 = 1 (10 = 3*3 + 1)

20 % 4 = 0 (20 = 5*4 + 0)

23 % 4 = 3 (23 = 5*4 + 3)

24 % 4 = 0 (24 = 6*4 + 0)

```
print(10%3)
print(20%4)
print(23%4)
print(24%4)
```

The last thing we learnt was if-else. We can for example use the following if-else to check if variable x is 0 or not. 

**REMEMBER TO USE == INSTEAD OF = IN IF-ELSE**

```
print("when x is 0, the if-else gives us:")
x = 0

if(x == 0):
  print("x is 0")
else:
  print("x is not 0")

print("------------------------------------------------------")

print("when x is 1, the if-else gives us:")
x = 1

if(x == 0):
  print("x is 0")
else:
  print("x is not 0")
```
