# Modulo

Now we introduce a new concept: Modulo. The operation of modulo is %. 

A modulo B gives the remainder of A / B.

For example 20 % 7 = 6, because 20 = 2 * 7 + 6, where the remainder is 6.

Another example is 15 % 5 = 0, because 15 = 5 * 3 + 0, 15 is divisible by 5 so there is no remainder.

# Condition (if-else)

The last thing we are going to learn in this notebook is If-Else. The basic structure is the following:



```
if(condition):
#TAB  DO SOMETHING
else:
#TAB  DO SOMETHING ELSE
```
For example, we want the computer to tell us if a variable x equals to 1. We do the following:


````
x = 1

if(x == 1):
  print("x is 1")
else:
  print("x is not 1")
````

Note that we use == instead of = in the condition. This is because to the computer, x = 1 means assigning 1 to x, and x == 1 checks if x is 1, they have different meanings.

Given a number `x`, check if it is a multiple of 3, print `yes`, otherwise print `no`. 

````
x = 101

#now suppose I want to check if a number can be expressed by 4n+2 (e.g. 2,6,10,14, 18)
if(x%4 == 2):
  print("yes")
else:
  print("no")
````

# Condition (elif)

If we want more than 2 conditions in one branch, we use `elif` (stands for else if) One example usage is:


```
if(x > 0):
  print('x is positive')
elif(x < 0):
  print('x is negative')
else:
  print('x is zero')
```
