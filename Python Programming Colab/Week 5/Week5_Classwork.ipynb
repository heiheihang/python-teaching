## Prime Factorization of Numbers

You may realize every integer can be "broken down" in to a product of prime factors.

For example 72 = 2 * 2 * 2 * 3 * 3

Also notice that every prime factorization of an integer is unique. 

How do we produce such factorisation for each integer?

We first would like to know how do we find all the factors of an integer first:

```
x = 72 

t = 1

factors = []

while(t <= x):
  if(x % t == 0):
    factors.append(t)
  t += 1

print(factors)
```
